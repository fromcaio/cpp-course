## 1. The C Struggle: Naming Fatigue

In C, every function must have a unique name.
```c
void print_int(int x);
void print_float(float x);
void print_string(char* x);
```

In C++, we can use **Function Overloading**. We can have multiple functions with the same name, as long as their **parameters** (signature) are different.


````c++
#include <iostream>
#include <string>

// Same name, different arguments
void print(int i) {
    std::cout << "Integer: " << i << std::endl;
}

void print(double d) {
    std::cout << "Double:  " << d << std::endl;
}

void print(std::string s) {
    std::cout << "String:  " << s << std::endl;
}

{
    print(42);      // Calls print(int)
    print(3.14);    // Calls print(double)
    print("Hello"); // Calls print(string) (implicit conversion from const char*)
}
````

## 2. Default Arguments

In C, if you want an optional parameter, you often pass `0` or `NULL` or write a wrapper function.
In C++, you can specify default values in the declaration.

**Rule:** Default arguments must be at the **end** of the parameter list.

````c++
// 'port' defaults to 8080 if not provided
void connect_to_server(std::string ip, int port = 8080) {
    std::cout << "Connecting to " << ip << ":" << port << "..." << std::endl;
}

{
    connect_to_server("192.168.1.1");       // Uses default port
    connect_to_server("127.0.0.1", 3000);   // Uses provided port
}
````

Overloading + Defaults can sometimes confuse the compiler.

```cpp
void func(int x);
void func(int x, int y = 0);

func(10); // ERROR! Which one? 
// Could be func(10) OR func(10, 0)
```

Always design your overloads to be distinct.
