## 1. The C Problem: `void* context`

In C, when using callbacks (like `pthread_create` or `qsort`), you often need to pass extra data. You end up casting structs to `void*`.

In C++, we use **Lambdas**. A Lambda is an anonymous function that can "capture" variables from its surrounding scope.

## 2. Basic Syntax

```cpp
[captures] (parameters) -> return_type {
    body
}
```

````c++
#include <iostream>
#include <vector>
#include <algorithm>

{
    auto say_hello = []() {
        std::cout << "Hello from Lambda!" << std::endl;
    };

    say_hello();
    
    // Direct execution
    [](int x) { std::cout << "x squared: " << x*x << std::endl; }(5);
}
````

## 3. Captures: The Killer Feature

This is what makes Lambdas powerful. They can see local variables.

* `[=]`: Capture everything by **Value** (Copy).
* `[&]`: Capture everything by **Reference** (Dangerous if lambda outlives the scope!).
* `[x]`: Capture only `x` by value.
* `[&x]`: Capture only `x` by reference.

````c++
{
    int multiplier = 10;
    int offset = 5;

    // We want to pass this logic to a function, but it needs 'multiplier' and 'offset'.
    // In C, we'd need a global variable or a void* struct.
    // In C++, we just capture them.

    auto calculate = [multiplier, offset](int input) {
        return (input * multiplier) + offset;
    };

    std::cout << "Result: " << calculate(2) << std::endl; // (2 * 10) + 5 = 25
}
````

## 4. Mutable Lambdas

By default, captures by value `[x]` are read-only (const) inside the lambda. To modify them (locally inside the lambda object), use `mutable`.

````c++
{
    int counter = 0;

    // Capture by value, but allow modification of the internal copy
    auto count_up = [counter]() mutable {
        counter++;
        std::cout << "Internal count: " << counter << std::endl;
    };

    count_up(); // 1
    count_up(); // 2

    std::cout << "External counter: " << counter << " (Unchanged)" << std::endl;
}
````
